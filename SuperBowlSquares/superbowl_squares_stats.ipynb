## Imports and Read Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv", index_col=False)
# increase index to start with 1 to represent Super Bowl number
df.index = range(1,len(df)+1)
df.tail(5)
```

## Create Boards - AFC on x-axis, NFC on y-axis

```python
# create empty board for each quarter
q1_sum = pd.DataFrame(0, index=range(10), columns=range(10))
q2_sum = pd.DataFrame(0, index=range(10), columns=range(10))
q3_sum = pd.DataFrame(0, index=range(10), columns=range(10))
q4_sum = pd.DataFrame(0, index=range(10), columns=range(10))

# populate quarter summary dataFrames and total summary df
for i in range(len(df["AFC"])):
    q1_sum[df["AFC Q1"].iloc[i]][df["NFC Q1"].iloc[i]] += 1
    q2_sum[df["AFC Q2"].iloc[i]][df["NFC Q2"].iloc[i]] += 1
    q3_sum[df["AFC Q3"].iloc[i]][df["NFC Q3"].iloc[i]] += 1
    q4_sum[df["AFC Q4"].iloc[i]][df["NFC Q4"].iloc[i]] += 1

total_sum = (q1_sum + q2_sum + q3_sum + q4_sum)   
```

## Calculate expected values
**Assuming \\$5 per square with 25\% percent payout at the end of each quarter: $125 per quarter**

```python
# ev = (occurances/total_occurances)*payout
q1_ev = q1_sum/len(df["AFC"]) * 125
q2_ev = q2_sum/len(df["AFC"]) * 125
q3_ev = q3_sum/len(df["AFC"]) * 125
q4_ev = q4_sum/len(df["AFC"]) * 125
total_ev = q1_ev + q2_ev + q3_ev + q4_ev
```

## How many square have never won?

```python
zeroes_by_col = (total_sum == 0).sum()
zeroes = zeroes_by_col.sum()
print(f"{zeroes} squares have never won a single quarter") 
```

## Which squares have never won?

```python
zero_sum = pd.DataFrame(0, index=range(10), columns=range(10))
for i in range(10):
    for j in range(10):
        if(total_sum[i][j] == 0):
            zero_sum[i][j] += 1

# show heatmap - 1 represents winless squares
plt.figure(figsize = (10,10))
sns.heatmap(zero_sum, annot=True, square=True, cbar=False)
None
```

## Which pairs of numbers have never won?
Ex: [2,5], and [5,2] have both never won

```python
zero_sum = pd.DataFrame(0, index=range(10), columns=range(10))
pairs = []
for i in range(10):
    for j in range(i,10):
        if((total_sum[i][j] == 0) and (total_sum[j][i] == 0)):
            zero_sum[i][j] += 1
            zero_sum[j][i] += 1
            pairs.append([i,j])

# show heatmap - 1 represents winless squares
plt.figure(figsize = (10,10))
sns.heatmap(zero_sum, annot=True, square=True, cbar=False)

print("Here are the squares where regardless of order they have never won")
print(pairs)
```

## How many squares have an above average expected value?

```python
above_avg_by_col = (total_ev > 10).sum()
above_avg = above_avg_by_col.sum()
print(f"{above_avg} squares have an above average expected value")  

count = 0
above_avg_df = pd.DataFrame(0, index=range(10), columns=range(10))
for i in range(10):
    for j in range(10):
        if(total_ev[i][j] > 10):
            above_avg_df[i][j] += total_ev[i][j]
            count += total_ev[i][j]
            
count_percent = round(100*count/500,2)
print(f"The sum of their expected values is: ${round(count,2)}, which is {count_percent}% of the total winnings")

plt.figure(figsize = (10,10))
sns.heatmap(above_avg_df, annot=True, square=True)
None
```

### Create function to print out expected value for given square, with a given payout

```python
def square_ev(squares, payout, quarter_sum):
    # create expected for each quarter based on payout provided
    num_sbs = 55
    q1_ev = (quarter_sum[0]/num_sbs) * payout[0]
    q2_ev = (quarter_sum[1]/num_sbs) * payout[1]
    q3_ev = (quarter_sum[2]/num_sbs) * payout[2]
    q4_ev = (quarter_sum[3]/num_sbs) * payout[3]
    total_ev = q1_ev + q2_ev + q3_ev + q4_ev
    
    # calcualte excpected values of squares
    my_ev = 0
    for i in range(len(squares)):
        afc = squares[i][0]
        nfc = squares[i][1]
        my_ev += total_ev.loc[afc, nfc]
        
    wager = (sum(payout)/100)*len(squares)
    total = round(my_ev - wager, 2)
    
    if(my_ev > 0):
        print(f"Your squares have an expected winnings of ${total}")
    else:
        print(f"Your squares have an expected losses of -${total}")
```

```python
# list of quarter sums
quarter_list = [q1_sum, q2_sum, q3_sum, q4_sum]

# list of quarter payout totals
quarter_payouts = [250, 500, 250, 1000]
```

```python
# list of my squares [AFC, NFC]
my_squares = [[5,6], [6,3]]

# call function to get expected values
ev = square_ev(my_squares, quarter_payouts, quarter_list)
```
